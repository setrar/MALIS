The closed-form solution for a linear regression

$$
{\color{Green}\hat{w}} = ( {\color{Salmon}X}^T {\color{Salmon}X} )^{− 1} {\color{Salmon}X}^T {\color{Salmon}y}
$$

```python
import numpy as np
```

```python
y = np.array([[9, 8, 7],
              [6, 5, 4],
              [3, 2, 1]])
```

```python
y
```

```python
X = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
X
```

```python
X_transpose = X.T # np.transpose(X)
```

```python
X_transpose
```

```python
result = np.dot(X_transpose, X) # Xᵀ * X
```

```python
result
```

```python
inverse_result = np.linalg.inv(result) # (Xᵀ * X)⁻¹ 
```

```python
inverse_result
```

$$
{\color{Green}\hat{w}} = ( {\color{Salmon}X}^T {\color{Salmon}X} )^{− 1} {\color{Salmon}X}^T {\color{Salmon}y}
$$

```python
w_hat = np.linalg.inv(X_transpose @ X) @ X_transpose @ y
```

```python
w_hat
```

```python
w_hat[0,1]
```

```python
w_hat[1,1]
```

```python
import matplotlib.pyplot as plt
```

```python
w_hat.shape[0]
```

```python
# Get the number of column vectors
num_vectors = w_hat.shape[1]

# Plot each column vector on a separate line using a for loop
for i in range(num_vectors):
    plt.plot(w_hat[:, i], label=f'y{i + 1}')

plt.legend(loc='upper left')
plt.xlabel('Python Indexes are 0-based')

# Show the plot
plt.show()
```
