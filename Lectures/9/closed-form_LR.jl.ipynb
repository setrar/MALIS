The closed-form solution for a linear regression

$$
{\color{Green}\hat{w}} = ( {\color{Salmon}X}^T {\color{Salmon}X} )^{− 1} {\color{Salmon}X}^T {\color{Salmon}y}
$$

```julia
# Postfix Operations to simplify the reading
include("postfixes.jl")
```

```julia
y = [ 9 8 7 ; 6 5 4 ; 3 2 1 ]
```

```julia
X = [ 1 2 ; 3 4 ; 5 6 ]
```

```julia
Xᵀ
```

```julia
Xᵀ * X # --> Xᵀ = transpose(X)
```

```julia
(Xᵀ * X)⁻¹ # Matrix Inverse
```

$$
{\color{Green}\hat{w}} = ( {\color{Salmon}X}^T {\color{Salmon}X} )^{− 1} {\color{Salmon}X}^T {\color{Salmon}y}
$$

```julia
ŵ = (Xᵀ * X)⁻¹ * Xᵀ * y; @show ŵ;
```

```julia
round.(ŵ, digits=1) # Rounding the printout like Python
```

```julia
using Plots
plot(round.(ŵ, digits=1), xlabel="Julia Indexes are 1-based")
```
