The closed-form solution for a linear regression

$$
{\color{Green}\hat{w}} = ( {\color{Salmon}X}^T {\color{Salmon}X} )^{− 1} {\color{Salmon}X}^T {\color{Salmon}y}
$$

```julia
# Postfix Operations to simplify the reading
include("postfixes.jl")
```

```julia
y = [ 9 8 7 ; 6 5 4 ; 3 2 1 ]
```

```julia
X = [ 1 2 ; 4 5 ; 6 7 ]
```

```julia
Xᵀ * X # --> Xᵀ = transpose(X)
```

```julia
(Xᵀ * X)⁻¹ # Matrix Inverse
```

$$
(X^T * X)^{-1} * X^T
$$

```julia
result = (Xᵀ * X)⁻¹ * Xᵀ * y; @show result;
```

```julia
using Plots
plot((Xᵀ * X)⁻¹ * Xᵀ * y)
```
